```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Assuming df2 is already loaded
# Define features and target
features = ['Timestamp', 'weekday', 'festival_impact','Piligrims_7Day_Avg' ]
target = 'Piligrims'

# Train-test split
X = df2[features]
y = df2[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=3)

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Hyperparameter Tuning using GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

etr = ExtraTreesRegressor(random_state=3)
grid_search = GridSearchCV(etr, param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# Best model from Grid Search
best_etr = grid_search.best_estimator_

# Cross-Validation Score (5-Fold CV)
cv = KFold(n_splits=5, shuffle=True, random_state=3)
cv_scores = cross_val_score(best_etr, X_train_scaled, y_train, cv=cv, scoring='r2')
print(f"Cross-Validation R² Scores: {cv_scores}")
print(f"Mean CV R² Score: {np.mean(cv_scores):.4f}")

# Predictions
y_pred = best_etr.predict(X_test_scaled)

# Evaluation Metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Print evaluation results
print("\nModel Evaluation Metrics:")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R² Score: {r2:.2f}")

# Feature Importance Plot
importances = best_etr.feature_importances_
plt.figure(figsize=(8, 5))
plt.barh(features, importances, color='purple')
plt.xlabel("Feature Importance Score")
plt.ylabel("Features")
plt.title("Feature Importance in Extra Trees Regressor")
plt.show()
```

```python
import pandas as pd
import numpy as np
import time
from datetime import datetime, timedelta

# --- Input: Date for prediction ---
future_date_str = input("Enter the future date (YYYY-MM-DD): ")
future_date = datetime.strptime(future_date_str, "%Y-%m-%d")

# --- User inputs ---
is_weekend = int(input("Is it a weekend? (1 = Yes, 0 = No): "))
is_holiday = int(input("Is it a holiday? (1 = Yes, 0 = No): "))
festival_impact = int(input("Enter festival impact value (e.g., 65000): "))

# --- Convert date to UNIX timestamp ---
timestamp = int(time.mktime(future_date.timetuple()))

# --- Calculate 7-day average from previous years (2022, 2023, 2024) ---
pilgrims_avgs = []

for past_year in [2022, 2023, 2024]:
    try:
        past_date = future_date.replace(year=past_year)
        start_date = past_date - timedelta(days=7)
        start_ts = int(time.mktime(start_date.timetuple()))
        end_ts = int(time.mktime(past_date.timetuple()))

        past_data = df2[(df2["Timestamp"] >= start_ts) & (df2["Timestamp"] < end_ts)]
        avg = past_data["Piligrims"].mean()
        if not np.isnan(avg):
            pilgrims_avgs.append(avg)
    except ValueError:
        # Handles issues like Feb 29 not existing in non-leap years
        continue

# --- Final 7-day average ---
if pilgrims_avgs:
    pilgrims_7day_avg = np.mean(pilgrims_avgs)
else:
    pilgrims_7day_avg = df2["Piligrims"].mean()  # fallback

# --- Apply holiday adjustment (3.36% increase) ---
if is_holiday:
    pilgrims_7day_avg *= 1.0336

# --- Create input for prediction ---
new_data = {
    'Timestamp': timestamp,
    'weekday': is_weekend,
    'festival_impact': festival_impact,
    'Piligrims_7Day_Avg': pilgrims_7day_avg
}

new_data_df = pd.DataFrame([new_data])

# --- Scale and Predict ---
X_new_scaled = scaler.transform(new_data_df)
predicted_pilgrims = best_etr.predict(X_new_scaled)

# --- Output ---
print(f"Predicted Number of Pilgrims on {future_date_str}: {predicted_pilgrims[0]:,.0f}")
```

```python
import pandas as pd
import numpy as np
import time
from datetime import datetime, timedelta

# --- Input: Date for prediction ---
future_date_str = input("Enter the future date (YYYY-MM-DD): ")
future_date = datetime.strptime(future_date_str, "%Y-%m-%d")

# --- User inputs ---
is_weekend = int(input("Is it a weekend? (1 = Yes, 0 = No): "))
is_holiday = int(input("Is it a holiday? (1 = Yes, 0 = No): "))
user_festival = input("Enter the festival name: ")

# --- Convert date to UNIX timestamp ---
timestamp = int(time.mktime(future_date.timetuple()))

#calculate the festival impact
df = pd.read_csv("unpreprocesseddata.csv")

# Get impact
impact_value = get_festival_impact(user_festival, df)

# --- Calculate 7-day average from 2024 only ---
pilgrims_7day_avg = df2[
    (df2["Timestamp"] >= int(time.mktime((future_date.replace(year=2024) - timedelta(days=7)).timetuple()))) &
    (df2["Timestamp"] < int(time.mktime(future_date.replace(year=2024).timetuple())))
]["Piligrims"].mean()

# --- Fallback if data is missing ---
if np.isnan(pilgrims_7day_avg):
    pilgrims_7day_avg = df2["Piligrims"].mean()

# --- Apply holiday adjustment (3.36% increase) ---
if is_holiday:
    pilgrims_7day_avg *= 1.0336

# --- Create input for prediction ---
new_data = {
    'Timestamp': timestamp,
    'weekday': is_weekend,
    'festival_impact': impact_value,
    'Piligrims_7Day_Avg': pilgrims_7day_avg
}

new_data_df = pd.DataFrame([new_data])

# --- Scale and Predict ---
X_new_scaled = scaler.transform(new_data_df)
predicted_pilgrims = best_etr.predict(X_new_scaled)

# --- Output ---
print(f"Predicted Number of Pilgrims on {future_date_str}: {predicted_pilgrims[0]:,.0f}")
```

```python
def get_festival_impact(festival_name, df):
    """
    Returns the festival impact value for a given festival name
    using exponential smoothing-based encoding.
    
    Args:
        festival_name (str): Name of the festival.
        df (pd.DataFrame): DataFrame with 'Special Day' and 'Piligrims' columns.
        
    Returns:
        float: Encoded festival impact value.
    """
    # Calculate global mean
    global_mean = df['Piligrims'].mean()

    # Calculate raw festival means
    festival_means = df[df['Special Day'] != 'normal'].groupby('Special Day')['Piligrims'].mean()

    # Apply exponential smoothing
    smoothing_factor = 0.2
    if festival_name in festival_means:
        raw_mean = festival_means[festival_name]
        smoothed_impact = (1 - smoothing_factor) * raw_mean + smoothing_factor * global_mean
        return smoothed_impact
    else:
        # If festival not found, return smoothed global mean
        return global_mean
```
